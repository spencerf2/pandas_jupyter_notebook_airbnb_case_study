```python
airbnb = pd.read_csv('AB_NYC_2019.csv')
airbnb.head()
```

```python
# Which hosts are the busiest and why?

# Defining 'busiest' to mean 'have the most reviews_per_month across all their properties':

airbnb_copy = airbnb.copy()

print(f'These are the top 10 hosts with the most reviews each month across all their AirBnB rentals')
display(airbnb_copy.groupby(['host_id', 'host_name'])[['reviews_per_month']].sum().sort_values(by='reviews_per_month', ascending = False).head(10))

# Delete the below


# Defining 'busiest' to mean 'have the most guests staying with them across all listings in a year':
# (reviews_per_month * minimum_nights) * 12 is approximately equal to the number of visitors per year.
# airbnb_copy = airbnb.copy()

# airbnb_copy['num_visitors_per_year'] = ((airbnb_copy['reviews_per_month'] * airbnb_copy['minimum_nights']) * 12) / airbnb_copy['availability_365']

# Remove NaN:
# airbnb_copy['num_visitors_per_year'] = airbnb_copy['num_visitors_per_year'].fillna('N/A')
# airbnb_copy = airbnb_copy[airbnb_copy['num_visitors_per_year'] != 'N/A']

# display(airbnb_copy)
# Sum together totals for hosts.
# display(airbnb_copy.groupby(['host_id', 'host_name'])[['num_visitors_per_year']].sum().sort_values(by='num_visitors_per_year', ascending = False))
```

```python
# How many neighborhood groups are available and which shows up the most?

airbnb_copy = airbnb.copy()

neighborhood_groups = airbnb_copy['neighbourhood_group'].to_frame()
neighborhood_list = neighborhood_groups.drop_duplicates(subset='neighbourhood_group', ignore_index=True)
neighborhood_count = neighborhood_list['neighbourhood_group'].count()
print(f"There are {neighborhood_count} neighborhood groups available, and their counts are as follows:")

neighborhood_groups['count'] = 1
count_of_neighborhood_groups = neighborhood_groups.groupby(['neighbourhood_group'])['count'].count()
count_of_neighborhood_groups.to_frame()
```

```python
# Are private rooms the most popular in manhattan?
manhattan_df = airbnb.copy()

manhattan_df['count'] = 1

manhattan = manhattan_df[manhattan_df['neighbourhood_group'] == 'Manhattan']

print(f"No, 'Entire home/apt' is most popular:")

manhattan.groupby(['neighbourhood_group','room_type'])['count'].count().to_frame()
```

```python
# Which hosts are the busiest and based on their reviews?

print('These are the Top 10 Hosts with the most reviews (across all their properties):')

display(airbnb_copy.groupby(['host_id', 'host_name'])[['number_of_reviews']].sum().sort_values(by='number_of_reviews', ascending=False).head(10))
```

```python
#Which neighorhood group has the highest average price?

print('Manhattan has the highest average price per night.')

display(airbnb_copy.groupby(['neighbourhood_group'])[['price']].mean().sort_values(by='price', ascending=False).round(decimals=2))
```

```python
# Which neighbor hood group has the highest total price?

print('If all the AirBnBs in New York were rented, Manhattan owners, would make the most money in one night:')

display(airbnb_copy.groupby(['neighbourhood_group'])[['price']].sum().sort_values(by='price', ascending=False).round(decimals=2))
```

```python
#Which top 5 hosts have the highest total price?

airbnb_copy.groupby(['host_id', 'host_name'])[['price']].sum().sort_values(by='price', ascending=False).head(5)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

reviews_over_100 = airbnb_copy[airbnb_copy['number_of_reviews'] >= 100]
reviews_over_100_w_zero_availability = reviews_over_100[reviews_over_100['availability_365'] == 0]

print(f"The following properties have no availability and a review count of 100 or more:")
reviews_over_100_w_zero_availability
```

```python
# What host has the highest total of prices and where are they located?

print(f"The host with the highest total of prices is:")
display(airbnb_copy.groupby(['host_id', 'host_name'])[['price']].sum().sort_values(by='price', ascending=False).head(1))

print(f"They have the following properties:")
display(airbnb_copy[airbnb_copy['host_id'] == 219517861])
```

```python
# When did Danielle from Queens last receive a review?

queens = airbnb_copy[airbnb_copy['neighbourhood_group'] == 'Queens']

danielle_from_queens_properties = queens[queens['host_name'] == 'Danielle']

print(f"Danielle's last review was on:")
danielle_from_queens_properties[['last_review']].sort_values(by='last_review', ascending=False).head(1).reset_index(drop=True)
```

## Further Questions

1. Which host has the most listings?

```python
print(f"Sonder (NYC) has the most listings at 327:")

airbnb_copy.groupby(['host_id', 'host_name'])['id'].count().sort_values(ascending=False).head(1).to_frame()
```

2. How many listings have completely open availability?

```python
complete_availability = airbnb_copy.copy()
complete_availability['count'] = 1

total_fully_available = complete_availability[complete_availability['availability_365'] == 365]['count'].count()

print(f"There are {total_fully_available} listings that are available to book for any day of the year, or every"
      " day of the year.")
```

3. What room_types have the highest review numbers?

```python
print(f"Average number of reviews on each room type:")

airbnb_copy.groupby(['room_type'])['number_of_reviews'].mean().sort_values(ascending=False).to_frame()
```
